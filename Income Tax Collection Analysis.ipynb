# Below I will do the analysis of different trends under Direct Tax for the previous 15 years i.e., from period 2000-01 to 2015-16:

```python
# importing python packages required for the analysis:
import pandas as pd
import xlrd
import matplotlib.pyplot as plt
```

```python
# importing of data from data.gov.in in xls format for futher analysis:
loc = 'https://github.com/pshivashish/pshivashish/blob/main/datafile.xls?raw=true'
data = pd.read_excel(loc)
print(data)
```

```python
# ploting the trend of yearly Income Tax collection over the years:
x = list(data['Financial Year'])
y = list(data['Revenue Realization From - Income Tax'])

# plotting a line plot after changing it's width and height
f = plt.figure()
f.set_figwidth(15)
f.set_figheight(10)

plt.plot(x,y)

# naming the x axis
plt.xlabel('F.Y.')

# naming the y axis
plt.ylabel('Income Tax Collection in Cr. Rs.')

# giving a title to my graph
plt.title('Collection Trend In Income Tax')


# function to show the plot
plt.show()
```

```python
# Now we will see the percentage growth in Income Tax Collection from the previous year collection:

# ploting the trend of yearly Income Tax collection over the years:
a = list(data['Financial Year'])
b = list(data['% Annual Growth - Income Tax'])

# plotting a line plot after changing it's width and height
f = plt.figure()
f.set_figwidth(15)
f.set_figheight(10)

plt.plot(a,b)

# naming the x axis
plt.xlabel('F.Y.')

# naming the y axis
plt.ylabel('% Growth in Income Tax Collection')

# giving a title to my graph
plt.title('Collection Trend In Income Tax')


# function to show the plot
plt.show()
```
